```python
def time_slot(hour):
    if 5 <= hour and hour <=12:
        return 'Morning'
    elif 12 < hour and hour <= 17:
        return 'Afternoon'
    elif 17 < hour and hour <= 21:
        return 'Evening'
    else:
        return 'Night'
    
df['time_segment'] = df['hour'].apply(time_slot)
df['time_segment'].head(10)
```
